**Calculadora simples**

```python
print('Bem-vindo à calculadora simples!')

# Loop para garantir que o usuário insira dois números diferentes
while True:
    try:
        num1 = float(input('Digite o primeiro número: '))
        num2 = float(input('Digite o segundo número: '))
        if num1 == num2:
            print('Os números devem ser diferentes. Por favor, tente novamente.')
        else:
            break
    except ValueError:
        print('Entrada inválida. Por favor, digite apenas números.')

print(f'Você digitou: {num1} e {num2}')
```

```python
# Loop para garantir que o usuário escolha uma operação válida
while True:
    operacao = input('Escolha a operação (adicao ou subtracao): ').lower()
    if operacao in ['adicao', 'subtracao']:
        break
    else:
        print('Operação inválida. Por favor, escolha \'adicao\' ou \'subtracao\'.')

print(f'Você escolheu: {operacao}')
```

```python
# Realiza a operação escolhida e exibe o resultado dela (opcional, para contexto)
if operacao == 'adicao':
    resultado_operacao = num1 + num2
    print(f'O resultado da operação de adição é: {resultado_operacao}')
elif operacao == 'subtracao':
    resultado_operacao = num1 - num2
    print(f'O resultado da operação de subtração é: {resultado_operacao}')

# Conforme solicitado, mostra no final o resultado da soma dos dois números
resultado_soma_final = num1 + num2
print(f'No final, a soma dos dois números ({num1} + {num2}) é: {resultado_soma_final}')
```
